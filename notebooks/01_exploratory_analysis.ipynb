#### Check missing values

```python
df.isnull().sum()
```

```python
# Replace missing values with mean or median
# But since mean can get affected by outliers, it's better to use median
# Median is robust to outliers

# Do not fill all the missing ratings with the overall median rating
# Ratings can vary a lot depending on the category
# Clothing category can have different rating than footwear category

# Find the median values of each category, then replace the missing values of that category with the median
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Rename columns

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# Create age_group column

bins = [17, 29, 44, 59, 100]
labels = ["Young Adult", "Adult", "Middle-Aged", "Senior Adult"]

df["age_group"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels
)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Create purchase_frequency_days column

frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Every 3 Months': 90,
    'Quarterly': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
# Check if the two columns mean the same thing to reduce redundancy

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Remove one of the columns

df = df.drop('promo_code_used', axis=1)
```
